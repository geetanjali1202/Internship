Q4.Scrape data for first 100 sneakers you find when you visit flipkart.com and search for “sneakers” in the search field.
You have to scrape 3 attributes of each sneaker:
1. Brand
2. Product Description

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

#set webdriver
driver = webdriver.Chrome()

# enter url flipkart.com
driver.get("https://www.flipkart.com/")

# Find the search bar and enter "sneakers"
search_sneakers = driver.find_element(By.XPATH,"/html/body/div[1]/div/div[1]/div/div/div/div/div[1]/div/div[1]/div/div[1]/div[1]/header/div[1]/div[2]/form/div/div/input")
search_sneakers.send_keys('Sneakers')
search = search_sneakers = driver.find_element(By.XPATH,"/html/body/div[1]/div/div[1]/div/div/div/div/div[1]/div/div[1]/div/div[1]/div[1]/header/div[1]/div[2]/form/div/button")
search.click()

# Initialize lists to store scraped data
brands = []
descriptions = []
prices = []

# Loop through the first 100 sneakers
while len(brands) < 100:
    # Find all sneakers on the page
    sneaker_cards = driver.find_elements(By.XPATH, "//div[@class='_2WkVRV']")
    for card in sneaker_cards:
        # Extract brand
        brand = card.find_element(By.XPATH, '//div[@class="_2WkVRV"]').text
        brands.append(brand)
        
        # Extract description
        description = card.find_element(By.XPATH,'//a[@class="IRpwTa"]').text
        descriptions.append(description)
        
        # Extract price
        try:
            price = card.find_element(By.XPATH,'//a[@class="_3bPFwb"]').text
        except:
            price = "Price not available"
        prices.append(price)
        
        # Check if we've scraped 100 sneakers
        if len(brands) == 100:
            break

    # If there are more pages, navigate to the next page
    try:
        next_button = driver.find_element(By.XPATH, '//a[@class="_1LKTO3"]')
        next_button.click()
        time.sleep(2)  # Wait for the page to load
    except:
        break  # Exit loop if there are no more pages

# Close the WebDriver
driver.quit()

# Print the scraped data
for i in range(100):
    print("Sneaker", i+1)
    print("Brand:", brands[i])
    print("Description:", descriptions[i])
    print("Price:", prices[i])
    print()
```

Q7: Write a python program to display list of respected former Prime Ministers of India (i.e. Name, Born-Dead, Term of office, Remarks) from https://www.jagranjosh.com/general-knowledge/list-of-all-prime-ministers-of-india-1473165149-1
scrap the mentioned data and make the DataFrame

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
```

```python
driver = webdriver.Chrome()
url = "https://www.jagranjosh.com/general-knowledge/list-of-all-prime-ministers-of-india-1473165149-1"
driver.get(url)
```

```python
# Find the table containing the list of Prime Ministers
table = driver.find_element(By.CLASS_NAME, "TableData")
```

```python
# Initialize lists to store scraped data
names = []
born_dead = []
term_of_office = []
remarks = []
```

```python
# Loop through each row in the table
rows = table.find_elements(By.TAG_NAME, "tr")[1:]  # Skip the header row
for row in rows:
    # Extract data from each cell in the row
    cells = row.find_elements(By.TAG_NAME, "td")
    name = cells[0].text.strip()
    bd = cells[1].text.strip()
    to = cells[2].text.strip()
    remark = cells[3].text.strip()
    
    # Append data to lists
    names.append(name)
    born_dead.append(bd)
    term_of_office.append(to)
    remarks.append(remark)

# Create a DataFrame
data = {
    "Name": names,
    "Born-Dead": born_dead,
    "Term of Office": term_of_office,
    "Remarks": remarks
}

df = pd.DataFrame(data)

# Display the DataFrame
print(df)

# Close the WebDriver
driver.quit()
```
